```python
import deeplake
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
ds = deeplake.load('hub://activeloop/wiki-art')

# Initialize lists to store dimensions
heights = []
widths = []
channels = []

# Sample 1000 random images to analyze dimensions
for i in range(1000):
    img = ds[i]['images'].numpy()
    heights.append(img.shape[0])
    widths.append(img.shape[1]) 
    channels.append(img.shape[2])

# Create histograms
plt.figure(figsize=(15,5))

plt.subplot(131)
plt.hist(heights, bins=50)
plt.title('Distribution of Image Heights')
plt.xlabel('Height (pixels)')
plt.ylabel('Count')

plt.subplot(132)
plt.hist(widths, bins=50)
plt.title('Distribution of Image Widths') 
plt.xlabel('Width (pixels)')
plt.ylabel('Count')

plt.subplot(133)
plt.hist(channels, bins=5)
plt.title('Distribution of Color Channels')
plt.xlabel('Number of Channels')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

# Print summary statistics
print("\nSummary Statistics:")
print(f"Height - Mean: {np.mean(heights):.1f}, Min: {min(heights)}, Max: {max(heights)}")
print(f"Width - Mean: {np.mean(widths):.1f}, Min: {min(widths)}, Max: {max(widths)}")
print(f"Channels - Most common: {max(set(channels), key=channels.count)}")
```
